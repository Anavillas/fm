```python

# ## 1. Importação das Bibliotecas

import numpy as np
import matplotlib.pyplot as plt

# Configuração estética dos gráficos
plt.style.use('seaborn-v0_8-whitegrid')
```

```python

# ## 2. Definição de Funções Matemáticas

# Função de 1º grau
def f1(x, a=2, b=1):
    return a*x + b

# Função de 2º grau
def f2(x, a=1, b=-4, c=3):
    return a*x**2 + b*x + c

# Função polinomial (exemplo para IA — aproximação não linear)
def f3(x):
    return 0.5*x**3 - 3*x + 1

# Função de ativação (comum em IA)
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

```

```python
# ## 3. Intervalo de valores para análise

x = np.linspace(-10, 10, 400)

# ## 4. Plot 1 — Funções polinomiais e interpretação geométrica

y1 = f1(x)
y2 = f2(x)
y3 = f3(x)

plt.figure(figsize=(10,5))
#plt.plot(x, y1, label="f1(x) = 2x + 1", color="blue")
#plt.plot(x, y2, label="f2(x) = x² - 4x + 3", color="red")
plt.plot(x, y3, label="f3(x) = 0.5x³ - 3x + 1", color="green")
plt.axhline(0, color="black", linewidth=0.8)
plt.axvline(0, color="black", linewidth=0.8)
plt.title("Comparação de Funções — Crescimento Linear, Quadrático e Cúbico")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.show()

# ---
# ### Interpretação:
# - f1(x): representa uma relação linear, típica de regressões simples.
# - f2(x): demonstra comportamento não-linear (parábola).
# - f3(x): aproxima relações mais complexas — ideia central em redes neurais profundas,
#   onde cada camada aplica uma função não-linear semelhante.
```

```python

# ## 5. Plot 2 — Funções de Ativação (IA)
# As funções de ativação definem como os neurônios artificiais "reagem" aos estímulos.

# ReLU (Rectified Linear Unit)
def relu(x):
    return np.maximum(0, x)

# Tangente hiperbólica
def tanh(x):
    return np.tanh(x)

# Cálculo das funções de ativação
y_sigmoid = sigmoid(x)
y_relu = relu(x)
y_tanh = tanh(x)

plt.figure(figsize=(10,5))
plt.plot(x, y_sigmoid, label="Sigmoid", color="blue")
plt.plot(x, y_relu, label="ReLU", color="orange")
plt.plot(x, y_tanh, label="tanh", color="green")
plt.title("Funções de Ativação em Inteligência Artificial")
plt.xlabel("Entrada (x)")
plt.ylabel("Saída f(x)")
plt.axhline(0, color="black", linewidth=0.8)
plt.axvline(0, color="black", linewidth=0.8)
plt.legend()
plt.show()

# ---
# ### Interpretação:
# - **Sigmoid:** suaviza valores e é usada para decisões binárias (entre 0 e 1).
# - **ReLU:** ativa apenas valores positivos — muito usada em redes neurais modernas.
# - **tanh:** centraliza valores entre -1 e 1, útil para normalização em camadas ocultas.

```

 
9. Determine se f(x) = 3x + 2 é crescente ou decrescente.  


```python
coef = [3, 2]

# Derivada usando numpy
derivada = np.polyder(coef)

print("Derivada:", derivada)

if derivada[0] > 0:
    print("A função é crescente.")
elif derivada[0] < 0:
    print("A função é decrescente.")
else:
    print("A função é constante.")
```


12. Encontre f(-2) para f(x) = x² + 3x - 4.  


```python
import numpy as np
import matplotlib.pyplot as plt

# Coeficientes da função f(x) = x^2 + 3x - 4
coef = [1, 3, -4]

# Valor para calcular f(-2)
x_value = -2

# Usando numpy para calcular f(-2)
resultado = np.polyval(coef, x_value)

print(f"f(-2) = {resultado}")

# Criar valores para o gráfico
x = np.linspace(-5, 5, 100)
y = np.polyval(coef, x)

# Plot do gráfico
plt.figure()
plt.plot(x, y, label="f(x) = x² + 3x - 4")
plt.scatter(x_value, resultado, label=f"Ponto f(-2) = {resultado}", s=100)
plt.title("Gráfico de f(x) = x² + 3x - 4 com ponto f(-2) destacado")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend()
plt.show()
```

```python
a = 2
b = -8
c = 6

x_v = -b / (2 *a)
y_v = np.polyval([a,b,c], x_v)

print(f"Vértice: ({x_v}), {y_v}")
```

```python


# Coeficientes da função f(x) = 2x² - 8x + 6
a = 2
b = -8
c = 6

# Cálculo do vértice
x_v = -b / (2 * a)
coeficientes = [a, b, c]
y_v = np.polyval(coeficientes, x_v)

print(f"O vértice da parábola é: ({x_v}, {y_v})")

# Criar valores para o gráfico
x = np.linspace(-1, 5, 200)
y = np.polyval(coeficientes, x)

# Plotar o gráfico
plt.figure()
plt.plot(x, y, label="f(x) = 2x² - 8x + 6")
plt.scatter([x_v], [y_v], s=100, label=f"Vértice ({x_v}, {y_v})")
plt.title("Gráfico de f(x) = 2x² - 8x + 6 com vértice destacado")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend()
plt.show()
```


15. Verifique se f(x) = x² - 9 corta o eixo x.  


```python
import numpy as np
import matplotlib.pyplot as plt

# Coeficientes da função f(x) = x^2 - 9
coeficientes = [1, 0, -9]

# Encontrar raízes com NumPy
raizes = np.roots(coeficientes)
print("As raízes da função são:", raizes)
print("Isso significa que a função corta o eixo x nesses pontos.")

# Criar valores para o gráfico
x = np.linspace(-5, 5, 200)
y = np.polyval(coeficientes, x)

# Plotar o gráfico
plt.figure()
plt.plot(x, y, label="f(x) = x² - 9")
plt.axhline(0, color='black', linewidth=1)  # eixo x
plt.scatter(raizes, [0, 0], label=f"Cortes no eixo x: {raizes}", s=100)
plt.title("Gráfico de f(x) = x² - 9 com interseções no eixo x")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend()
plt.show()

```


16. Determine o número de raízes reais de f(x) = x² + 4x + 9.  


```python
import numpy as np

# Coeficientes da função f(x) = x^2 + 4x + 9
coeficientes = [1, 4, 9]

# Usando numpy.roots para encontrar raízes
raizes = np.roots(coeficientes)

print("Raízes encontradas:", raizes.round(2))

# Verificando se são reais
if np.iscomplex(raizes).any():
    print("A função não possui raízes reais.")
else:
    print("A função possui raízes reais.")
```
